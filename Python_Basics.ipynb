## None

`None` is a special value in Python that represents the absence of a value. It is commonly used as a default value for function arguments and to indicate that a variable has no value yet.

### Iterating Through Dictionaries

There are several ways to loop through a dictionary:

````
student = {'name': 'Alice', 'age': 22, 'major': 'Computer Science'}

# Iterate through keys (default behavior)
print("Keys:")
for key in student:
    print(f"  {key}")

# Iterate through values
print("\nValues:")
for value in student.values():
    print(f"  {value}")

# Iterate through key-value pairs (most common)
print("\nKey-Value pairs:")
for key, value in student.items():
    print(f"  {key}: {value}")
````

### The get() Method

Use `get()` to safely access dictionary values without risking a KeyError:

````
student = {'name': 'Alice', 'age': 22}

# Using get() with a default value
print(student.get('name'))           # Returns 'Alice'
print(student.get('gpa'))            # Returns None (key doesn't exist)
print(student.get('gpa', 0.0))       # Returns 0.0 (default value)

# Compare with direct access:
# print(student['gpa'])  # This would raise KeyError!
````

## Tuple

A tuple is similar to a list, but it is **immutable** (cannot be changed after creation). Tuples use parentheses `()` instead of square brackets `[]`.

# while loops

A `while` loop continues executing as long as its condition is `True`. Be careful to ensure the condition eventually becomes `False`, or you'll create an infinite loop!

````
count = 0

while count < 5:
    print(f"Count is {count}")
    count += 1  # Don't forget to update the condition!

print("Loop finished!")
````

### break and continue

- `break` exits the loop immediately
- `continue` skips to the next iteration

````
# Using break to exit early
for i in range(10):
    if i == 5:
        print("Found 5, stopping!")
        break
    print(i)

print("---")

# Using continue to skip iterations
for i in range(5):
    if i == 2:
        continue  # Skip printing 2
    print(i)
````

## Exceptions (`try` / `except`)

When Python hits an error (an *exception*), execution stops unless you handle it. For data work, it’s common to catch expected exceptions (like converting a non-numeric string to an integer).

Good practice:
- Catch *specific* exceptions (e.g., `ValueError`) rather than a bare `except:`.
- Keep the `try` block small—only wrap the lines that might fail.


````
values = ['10', 'oops', '30']
clean = []

for v in values:
    try:
        clean.append(int(v))
    except ValueError:
        clean.append(None)

clean
````

You can also raise exceptions intentionally when validating inputs. This is useful when writing helper functions for assignments.


````
def positive_sqrt(x):
    """Return sqrt(x) for x >= 0; raise a helpful error otherwise."""
    if x < 0:
        raise ValueError('x must be non-negative')
    return math.sqrt(x)

positive_sqrt(9)
````

### Default Parameter Values

You can give parameters default values, making them optional when calling the function:

````
def power(base, exponent=2):
    """Calculate base raised to exponent. Default is squaring."""
    return base ** exponent

print(power(5))      # Uses default exponent=2, returns 25
print(power(5, 3))   # Uses exponent=3, returns 125
print(power(2, 10))  # Returns 1024
````

### Positional vs Keyword Arguments

When you call a function, Python matches the values you pass to the function's **parameters** in two main ways:

- **Positional arguments**: matched **by order**
- **Keyword arguments**: matched **by parameter name**

Keyword arguments often make code easier to read and reduce mistakes. You'll see this style a lot in deep learning libraries (for example, `activation="relu"` in Keras).

**Rules to remember**
- Positional arguments come first; keyword arguments come after.
- With keyword arguments, the order does not matter.
- You cannot provide the same parameter twice (once positionally and once by keyword).


````
def greet(name, greeting="Hello"):
    print(greeting, name)

# Positional arguments (matched by order)
greet("Alice")
greet("Alice", "Hi")

# Keyword arguments (matched by name)
greet(name="Alice")
greet(name="Alice", greeting="Hi")
greet(greeting="Hi", name="Alice")  # order does not matter

# Mixing positional + keyword (very common)
greet("Alice", greeting="Howdy")

# Common mistake: providing the same parameter twice
# (Try it to see the error message)
# greet("Alice", name="Bob")

# Deep learning tie-in (you'll see calls like this in Keras/TensorFlow):
# Dense(64, activation="relu")
# Dense(units=64, activation="relu")  # equivalent to the line above
````

### *args and **kwargs

These special syntaxes allow functions to accept a variable number of arguments:
- `*args` collects extra positional arguments as a tuple
- `**kwargs` collects extra keyword arguments as a dictionary

````
def print_all(*args):
    """Print any number of arguments."""
    for arg in args:
        print(arg)

print_all("Hello", "World", 123)
print("---")

def print_info(**kwargs):
    """Print key-value pairs."""
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=25, city="Dallas")
````

We can use `lambda` functions to create small functions in Python.

A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.

Let's say we want to write a shorter version of the following function:

````
def square(number):
    return number*number

y = square(5)
print(y)
````

The shorter version is written like this:

````
square = lambda number: number * number

y = square(5)
print(y)
````

# Classes and Objects

Python is an object-oriented programming language. Understanding classes and objects is essential for working with deep learning frameworks like Keras and TensorFlow, where models, layers, and callbacks are all implemented as classes.

A **class** is a blueprint for creating objects. An **object** is an instance of a class.

## Defining a Simple Class

Use the `class` keyword to define a class. The `__init__` method is a special method called the **constructor** - it runs automatically when you create a new object.

````
class Dog:
    def __init__(self, name, age):
        self.name = name  # instance attribute
        self.age = age    # instance attribute

    def bark(self):
        print(f"{self.name} says: Woof!")

    def describe(self):
        print(f"{self.name} is {self.age} years old.")
````

## Creating Objects (Instances)

Once a class is defined, you can create objects by calling the class like a function.

````
# Create two Dog objects
my_dog = Dog("Buddy", 3)
your_dog = Dog("Luna", 5)

# Access attributes
print(my_dog.name)
print(your_dog.age)

# Call methods
my_dog.bark()
your_dog.describe()
````

## The `self` Parameter

The `self` parameter refers to the current instance of the class. It allows you to access the object's attributes and methods from within the class.

**Important:** `self` must be the first parameter of every instance method, but you don't pass it when calling the method - Python handles this automatically.

````
class Counter:
    def __init__(self):
        self.count = 0

    def increment(self):
        self.count += 1

    def get_count(self):
        return self.count

c = Counter()
c.increment()
c.increment()
c.increment()
print(c.get_count())  # Output: 3
````

## Class Attributes vs Instance Attributes

- **Instance attributes** are unique to each object (defined in `__init__` using `self.`)
- **Class attributes** are shared by all instances of the class

````
class Student:
    school = "UTA"  # Class attribute - shared by all students

    def __init__(self, name, major):
        self.name = name    # Instance attribute
        self.major = major  # Instance attribute

student1 = Student("Alice", "Computer Science")
student2 = Student("Bob", "Data Science")

# Both share the same school
print(student1.school)  # UTA
print(student2.school)  # UTA

# But have different names and majors
print(student1.name, student1.major)
print(student2.name, student2.major)
````

## Inheritance

Inheritance allows a class to inherit attributes and methods from another class. This is a fundamental concept in deep learning frameworks:

- In Keras, custom layers inherit from `keras.layers.Layer`
- Custom models inherit from `keras.Model`
- Custom callbacks inherit from `keras.callbacks.Callback`

````
# Parent class (base class)
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass  # To be implemented by subclasses

# Child class (derived class)
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name)  # Call parent's __init__
        self.color = color

    def speak(self):
        return f"{self.name} says Meow!"

# Another child class
class Bird(Animal):
    def speak(self):
        return f"{self.name} says Tweet!"

cat = Cat("Whiskers", "orange")
bird = Bird("Tweety")

print(cat.speak())
print(bird.speak())
print(f"{cat.name} is {cat.color}")
````

## The `super()` Function

The `super()` function is used to call methods from the parent class. This is especially important in deep learning when you need to extend the functionality of existing layers or models while preserving their original behavior.

````
class Shape:
    def __init__(self, color):
        self.color = color

    def describe(self):
        return f"A {self.color} shape"

class Rectangle(Shape):
    def __init__(self, color, width, height):
        super().__init__(color)  # Initialize parent class
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def describe(self):
        # Extend parent's describe method
        parent_desc = super().describe()
        return f"{parent_desc} - Rectangle ({self.width}x{self.height})"

rect = Rectangle("blue", 5, 3)
print(rect.describe())
print(f"Area: {rect.area()}")
````

## Special Methods (Dunder Methods)

Python classes can implement special methods (also called "dunder" methods for their double underscores) to customize object behavior. These are commonly used in deep learning frameworks.

````
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        """Called by print() and str()"""
        return f"Vector({self.x}, {self.y})"

    def __repr__(self):
        """Official string representation"""
        return f"Vector({self.x}, {self.y})"

    def __add__(self, other):
        """Allows using + operator"""
        return Vector(self.x + other.x, self.y + other.y)

    def __len__(self):
        """Called by len()"""
        return 2  # A 2D vector has 2 components

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2  # Uses __add__

print(v1)       # Uses __str__
print(v3)
print(len(v1))  # Uses __len__
````

## The `__call__` Method

The `__call__` method allows an object to be called like a function. This is heavily used in deep learning frameworks - for example, when you "call" a Keras layer or model, you're actually invoking its `__call__` method.

````
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, x):
        return x * self.factor

double = Multiplier(2)
triple = Multiplier(3)

# Use objects like functions!
print(double(5))   # Output: 10
print(triple(5))   # Output: 15
print(double(10))  # Output: 20
````

## Practical Example: A Simple Neural Network Layer

This example demonstrates how classes are used in deep learning. While this is a simplified version, it shows the same patterns used in actual frameworks like Keras.

````
class SimpleLayer:
    """A simplified dense layer to demonstrate OOP concepts in deep learning"""

    def __init__(self, input_size, output_size, activation=None):
        self.input_size = input_size
        self.output_size = output_size
        self.activation = activation
        # In real implementations, weights would be initialized here
        self.weights = None
        self.bias = None

    def __call__(self, inputs):
        """Forward pass - this is called when you use the layer"""
        # Simplified: just demonstrate the pattern
        print(f"Processing input of shape: {len(inputs)} elements")
        print(f"Output will have {self.output_size} elements")
        if self.activation:
            print(f"Applying {self.activation} activation")
        return [0] * self.output_size  # Placeholder output

    def __repr__(self):
        return f"SimpleLayer({self.input_size} -> {self.output_size}, activation={self.activation})"

# Create layers (similar to Keras)
layer1 = SimpleLayer(784, 128, activation='relu')
layer2 = SimpleLayer(128, 10, activation='softmax')

print(layer1)
print(layer2)
print()

# Use the layer like a function
dummy_input = [0.5] * 784
output = layer1(dummy_input)
````

````
# Basic type hints
def greet(name: str) -> str:
    return f"Hello, {name}!"

def add_numbers(a: int, b: int) -> int:
    return a + b

# Type hints with lists and dictionaries
from typing import List, Dict

def process_scores(scores: List[int]) -> float:
    return sum(scores) / len(scores)

def get_user_info() -> Dict[str, str]:
    return {"name": "Alice", "email": "alice@example.com"}

# Example usage
print(greet("World"))
print(add_numbers(5, 3))
print(process_scores([85, 90, 78, 92]))
````
